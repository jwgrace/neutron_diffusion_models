```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
import os
import sys
import subprocess
from glob import glob
from ipywidgets import interactive
import time

home_directory = os.getenv('HOME')
dStar_directory = home_directory + '/dStar/'
sys.path.append(dStar_directory + 'tools/')

import reader
```

```python
def load_LOGS_data(LOGS_directory):
    
    f = reader.dStarReader('{}/history.data'.format(LOGS_directory))

    # time after outburts in days
    t = f.data.time
    # log(L) [ergs]
    lg_L = f.data.lg_Lsurf
    # log(T_eff) [K]
    lg_Teff = f.data.lg_Teff
    
    # constants in cgs
    c = 3e10
    G = 6.67e-8
    M_sun = 1.989e33

    M = M_sun*f.header.total_mass
    R = f.header.total_radius*1e5
    redshift_factor = (1 - 2*G*M/(R*c**2))**(-.5)
    
    Teff_inf = 10**(lg_Teff)/redshift_factor
    
    return t, lg_L, lg_Teff, Teff_inf
```

```python
def make_directory(directory_name):
    try:
        os.mkdir(directory_name)
    except FileExistsError:
        pass
```

```python
def write_inlist(base_inlist_name, new_inlist_name, parameter_names, parameter_values, LOGS_directory):

    base_inlist = open(base_inlist_name, 'r')
    new_inlist = open(new_inlist_name, 'w')

    base_inlist.seek(0)

    for line in base_inlist.readlines():

        parameter_line = False
        for i, parameter_name in enumerate(parameter_names):
            if line[:7+len(parameter_name)] == '    {} = '.format(parameter_name):
                if parameter_name == 'core_temperature':
                    new_inlist.write('    {} = {:.4e}\n'.format(parameter_name, parameter_values[i]))
                    parameter_line = True
                    break
                else:
                    new_inlist.write('    {} = {:.4f}\n'.format(parameter_name, parameter_values[i]))
                    parameter_line = True
                    break

        # If we just modified a parameter line, move to the next line.
        if parameter_line == True:
            continue
        
        # When we reach the output_directory assignment line, reassign to the new output_directory.
        elif line[:23] == '    output_directory = ':
            new_inlist.write("    output_directory = '{}'\n".format(LOGS_directory))

        # If it is not one of the lines to change, then copy it to each new inlist.
        else:
            new_inlist.write(line)
            
    base_inlist.close()
    new_inlist.close()
```

```python
def run_model(inlist, dStar_directory='/home/justin/dStar/'):
    output_file = open('LOGS/output', 'w')
    subprocess.run(['./run_dStar', '-D', dStar_directory, '-I', inlist], stdout=output_file)
    output_file.close()

    with open('LOGS/output', 'r') as f:
        chi2_line = f.readlines()[-1]

    chi2 = float(chi2_line[7:])
    
    return chi2
```

```python
base_inlist = 'inlist_base'

# Observation times in days after outburst.
t_obs = np.array([65.1,235.7,751.6,929.5,1500.5,1582.9,3039.7,5312.1])

# Boltzmann constant for converting temperature from eV to MK
k_B = 8.617e-5
# Observed effective temperature in MK. Initial data is in eV and is converted.
Teff_obs = np.array([104.6,89.5,76.4,73.8,71.7,70.3,64.5,64.4])/k_B*10**-6
# Error on effective temperature in MK. Initial data is in eV and is converted.
Teff_obs_sigma = np.array([1.3,1.03,1.8,1.9,1.4,1.9,1.8,1.2])/k_B*10**-6


# Names of parameters to vary in dStar
# Units are: Msun, km, K, dimensionless, MeV/nucleon
parameter_names = ['core_mass', 'core_radius', 'core_temperature', 'Qimp', 'Q_heating_shallow']
number_of_parameters = len(parameter_names)
# Start sampling from fit_lightcurve example in dStar directory
parameter_values_best_fit = np.array([1.38, 9.32, 9.35e7, 4.4, 1.36])
# The range from the best-fit values 
# Values chosen from standard deviations of best-fit values in Merritt et al. 2016
#parameter_deltas = np.array([0.1, 0.5, 2.5e6, 2.0, 0.18])
# Vary all parameters by 10% of their best-fit values
parameter_deltas = .1*parameter_values_best_fit

models_per_parameter = 11
parameter_ranges = np.zeros((number_of_parameters, models_per_parameter))

for i, parameter in enumerate(parameter_names):
    parameter_min = parameter_values_best_fit[i] - parameter_deltas[i]
    parameter_max = parameter_values_best_fit[i] + parameter_deltas[i]
    
    parameter_ranges[i] = np.linspace(parameter_min, parameter_max, models_per_parameter)
    
chi2_values = np.zeros_like(parameter_ranges)
```

```python
best_fit_LOGS = 'LOGS_best_fit'
best_fit_inlist = 'inlist_best_fit'

make_directory(best_fit_LOGS)
write_inlist(base_inlist, best_fit_inlist, parameter_names, parameter_values_best_fit, best_fit_LOGS)

best_fit_chi2 = run_model(best_fit_inlist)
```

```python
start = time.time()

for i, parameter_name in enumerate(parameter_names):
    # Make a LOGS directory for varying this parameter
    varying_parameter_LOGS = 'LOGS-vary_{}'.format(parameter_name)
    make_directory(varying_parameter_LOGS)
    # Save the array of parameter values for this parameter in the LOGS directory for later use.
    np.save('{}/{}_values'.format(varying_parameter_LOGS, parameter_name), parameter_ranges[i])
    
    for j, parameter_value in enumerate(parameter_ranges[i]):
        LOGS_directory = '{}/LOGS-{:03d}'.format(varying_parameter_LOGS, j)
        inlist = '{}/inlist-{:03d}'.format(LOGS_directory, j)
        make_directory(LOGS_directory)
        write_inlist(best_fit_inlist, inlist, [parameter_name], [parameter_value], LOGS_directory)
        
        chi2_values[i, j] = run_model(inlist)
        
end = time.time()
```

```python
# Plot lightcurves of models and observational data with error bars

fig = plt.figure()
fig.patch.set_facecolor('white')
fig.set_dpi(150)

plt.plot(t_best_fit, Teff_inf_best_fit, label='Best-Fit Values')
plt.errorbar(t_obs, Teff_obs, yerr=Teff_obs_sigma, fmt='.', label='Observational Data')
plt.xlabel('t (days)')
plt.ylabel(r'T$_{eff}$ (MK)')
plt.title('KS 1731-260 Lightcurves')
plt.legend(loc='upper right', fontsize=6)
plt.show()
```

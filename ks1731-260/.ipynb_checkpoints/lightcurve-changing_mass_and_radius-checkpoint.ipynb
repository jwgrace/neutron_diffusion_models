```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
import os
import sys
from glob import glob
from ipywidgets import interactive

home_directory = os.getenv('HOME')
dStar_directory = home_directory + '/dStar/'
sys.path.append(dStar_directory + 'tools/')

import reader
```

```python
def load_LOGS_data(directory_name):
    
    f = reader.dStarReader('{}history.data'.format(directory_name))

    # time after outburts in days
    t = f.data.time
    # log(L) [ergs]
    lg_L = f.data.lg_Lsurf
    # log(T_eff) [K]
    lg_Teff = f.data.lg_Teff
    
    return t, lg_L, lg_Teff
```

```python
# Load base model output
LOGS_directory_1 = 'LOGS/'
LOGS_directory_2 = 'LOGS2/'

t1, lg_L1, lg_Teff1 = load_LOGS_data(LOGS_directory_1)
t2, lg_L2, lg_Teff2 = load_LOGS_data(LOGS_directory_2)
```

```python
# constants in cgs
c = 3e10
G = 6.67e-8
M_sun = 1.989e33

# neutron star mass (g) and radius (cm) for the two models
M_ns1 = 1.6*M_sun
R_ns1 = 1.1e6
M_ns2 = 1.4*M_sun
R_ns2 = 1e6

# redshift factor (1+z) for each neutron star
redshift_factor1 = (1 - 2*G*M_ns1/(R_ns1*c**2))**(-.5)
redshift_factor2 = (1 - 2*G*M_ns2/(R_ns2*c**2))**(-.5)

# convert local Teff from dStar to MK observed at infinity
Teff_MK1 = 10**(lg_Teff1 - 6)
Teff_MK_inf1 = Teff_MK1/redshift_factor2

Teff_MK2 = 10**(lg_Teff2 - 6)
Teff_MK_inf2 = Teff_MK2/redshift_factor2
```

```python
# Load data from output file to get observational data and model data at the times of observation.
output_file1 = open('{}output'.format(LOGS_directory_1), 'r')
t_obs, Teff_model1, Teff_min_obs, Teff_max_obs = np.loadtxt(output_file, skiprows=4, max_rows=8, unpack=True)

output_file2 = open('{}output'.format(LOGS_directory_2), 'r')
t_obs, Teff_model2, Teff_min_obs, Teff_max_obs = np.loadtxt(output_file, skiprows=4, max_rows=8, unpack=True)

# convert observational k_B*Teff from eV to Teff in MK
k_B = 8.617e-5
Teff_obs = np.array([103.2,88.9,75.5,73.3,71.0,66.0,70.3,63.1])/k_B*10**-6
Teff_obs_sigma = np.array([1.7,1.3,2.2,2.3,1.8,4.5,2.1,2.1])/k_B*10**-6
```

```python
# Plot lightcurves of models and observational data with error bars

fig = plt.figure()
fig.patch.set_facecolor('white')
fig.set_dpi(150)

plt.plot(t1, Teff_MK_inf1, label=r'(M = {:.2f} M$_\odot$, R = {:.2f} km)'.format(M_ns1/M_sun, R_ns1*1e-5))
plt.plot(t2, Teff_MK_inf2, label=r'(M = {:.2f} M$_\odot$, R = {:.2f} km)'.format(M_ns2/M_sun, R_ns2*1e-5))
plt.errorbar(t_obs, Teff_obs, yerr=Teff_obs_sigma, fmt='.', label='Observational Data')
plt.xlabel('t (days)')
plt.ylabel(r'T$_{eff}$ (MK)')
plt.title('KS 1731-260 Lightcurves')
plt.legend(loc='upper right', fontsize=6)
plt.show()
```

```python
# Calculate chi2 for each data point
chi2_1 = ((Teff_model1 - Teff_obs)/Teff_obs_sigma)**2
chi2_2 = ((Teff_model2 - Teff_obs)/Teff_obs_sigma)**2

print('Model 1 chi^2 =', np.sum(chi2_1))
print('Model 2 chi^2 =', np.sum(chi2_2))
```

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(6, 5))
fig.patch.set_facecolor('white')
fig.set_dpi(150)

ax1.plot(t_obs, Teff_model1, '.', label=r'(M = {:.2f} M$_\odot$, R = {:.2f} km)'.format(M_ns1/M_sun, R_ns1*1e-5))
ax1.plot(t_obs, Teff_model2, '.', label=r'(M = {:.2f} M$_\odot$, R = {:.2f} km)'.format(M_ns2/M_sun, R_ns2*1e-5))
ax1.errorbar(t_obs, Teff_obs, yerr=Teff_obs_sigma, fmt='.', label='Observational Data')
ax1.set_title(r'T$_{eff}$')
ax1.set_xlabel('t (days)')
ax1.set_ylabel(r'T$_{eff}$ (MK)')
ax1.legend()

ax2.plot(t_obs, chi2_1, '.', label=r'(M = {:.2f} M$_\odot$, R = {:.2f} km)'.format(M_ns1/M_sun, R_ns1*1e-5))
ax2.plot(t_obs, chi2_2, '.', label=r'(M = {:.2f} M$_\odot$, R = {:.2f} km)'.format(M_ns2/M_sun, R_ns2*1e-5))
ax2.set_title(r'$\chi^2$')
ax2.set_xlabel('t (days)')
ax2.set_ylabel(r'$\chi^2$')

fig.suptitle('Comparison to Observational Data')
fig.tight_layout()
```

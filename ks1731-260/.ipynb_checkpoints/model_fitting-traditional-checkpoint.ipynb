```python
import numpy as np
import os
import sys
import subprocess
import time
import emcee
import corner
from multiprocessing import Pool
import matplotlib.pyplot as plt

home_directory = os.getenv('HOME')
dStar_directory = home_directory + '/dStar/'
sys.path.append(dStar_directory + 'tools/')

import reader
```

```python
def log_probability(parameter_values):
    parameter_minimum_values = np.array([1.0, 5., 8.35e7, 0.0, 0.0])
    parameter_maximum_values = np.array([2.8, 20., 935e8, 20.0, 20.0])
    if np.any(parameter_values <= parameter_minimum_values) or np.any(parameter_values >= parameter_maximum_values):
        return -np.inf
    else:
        # Run the model and save the output chi2
        core_mass = str(parameter_values[0])
        core_radius = str(parameter_values[1])
        core_temperature = str(parameter_values[2])
        Qimp = str(parameter_values[3])
        Q_heating_shallow = str(parameter_values[4])
        try:
            chi2 = float(subprocess.run(['./run_dStar', '-M', core_mass, '-R', core_radius, '-T', core_temperature, '-Q', Qimp, '-H', Q_heating_shallow], capture_output=True, text=True).stdout)
            return -chi2/2
        except:
            return -np.inf
```

```python
dStar_directory = '/home/justin/dStar/' # dStar directory needed to run dStar models
base_inlist = 'inlist_base' # Base inlist with the initial parameters. Will make the "current" inlist based on this each step.
inlist = 'inlist' # Name for the inlist to be used for each step.

# Observation times in days after outburst.
t_obs = np.array([65.1,235.7,751.6,929.5,1500.5,1582.9,3039.7,5312.1])

# Boltzmann constant for converting temperature from eV to MK
k_B = 8.617e-5
# Observed effective temperature in MK. Initial data is in eV and is converted.
Teff_obs = np.array([104.6,89.5,76.4,73.8,71.7,70.3,64.5,64.4])/k_B*10**-6
# Error on effective temperature in MK. Initial data is in eV and is converted.
Teff_obs_sigma = np.array([1.3,1.03,1.8,1.9,1.4,1.9,1.8,1.2])/k_B*10**-6

# Names of parameters to vary in dStar
# Units are: Msun, km, K, dimensionless, MeV/nucleon
parameter_names = ['core_mass', 'core_radius', 'core_temperature', 'Qimp', 'Q_heating_shallow']
number_of_parameters = len(parameter_names)
# Start sampling from fit_lightcurve example in dStar directory
parameter_values_best_fit = np.array([1.39, 9.9, 9.35e7, 4.2, 1.37])

parameter_values_initial = np.array([parameter_values_best_fit])
```

```python
nwalkers = 32
ndim = len(parameter_names)
number_of_steps = 10000

pos = parameter_values_initial + .01*parameter_values_initial*np.random.randn(nwalkers, number_of_parameters)

with Pool() as pool:
    sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, pool=pool)
    sampler.run_mcmc(pos, number_of_steps, progress=True)
```

```python
def load_LOGS_data(LOGS_directory):
    
    f = reader.dStarReader('{}/history.data'.format(LOGS_directory))

    # time after outburts in days
    t = f.data.time
    # log(L) [ergs]
    lg_L = f.data.lg_Lsurf
    # log(T_eff) [K]
    lg_Teff = f.data.lg_Teff
    
    # constants in cgs
    c = 3e10
    G = 6.67e-8
    M_sun = 1.989e33

    M = M_sun*f.header.total_mass
    R = f.header.total_radius*1e5
    redshift_factor = (1 - 2*G*M/(R*c**2))**(-.5)
    
    Teff_inf = 10**(lg_Teff)/redshift_factor
    
    return t, lg_L, lg_Teff, Teff_inf
```

```python
# Observation times in days after outburst.
t_obs = np.array([65.1,235.7,751.6,929.5,1500.5,1582.9,3039.7,5312.1])

# Boltzmann constant for converting temperature from eV to MK
k_B = 8.617e-5
# Observed effective temperature in MK. Initial data is in eV and is converted.
Teff_obs = np.array([104.6,89.5,76.4,73.8,71.7,70.3,64.5,64.4])/k_B*10**-6
# Error on effective temperature in MK. Initial data is in eV and is converted.
Teff_obs_sigma = np.array([1.3,1.03,1.8,1.9,1.4,1.9,1.8,1.2])/k_B*10**-6
```

```python
t_best_fit, lg_L_best_fit, lg_Teff_best_fit, Teff_inf_best_fit = load_LOGS_data('LOGS')
Teff_inf_best_fit *= 1e-6

# Plot lightcurves of models and observational data with error bars

time_mask = [t_best_fit > 0.]

fig = plt.figure()
fig.patch.set_facecolor('white')
fig.set_dpi(150)

plt.loglog(t_best_fit[time_mask], Teff_inf_best_fit[time_mask], label='Best-Fit Values')
plt.errorbar(t_obs, Teff_obs, yerr=Teff_obs_sigma, fmt='.', label='Observational Data')
plt.xlabel('t (days)')
plt.ylabel(r'T$_{eff}$ (MK)')
plt.title('KS 1731-260 Lightcurves')
plt.legend(loc='upper right', fontsize=6)
plt.show()
```

```python
t_best_fit, lg_L_best_fit, lg_Teff_best_fit, Teff_inf_best_fit = load_LOGS_data('LOGS')
Teff_inf_best_fit *= 1e-6

t_best_fit2, lg_L_best_fit2, lg_Teff_best_fit2, Teff_inf_best_fit2 = load_LOGS_data('LOGS3')
Teff_inf_best_fit2 *= 1e-6

time_mask = [t_best_fit > 0.]
time_mask2 = [t_best_fit2 > 0.]

fig = plt.figure()
fig.patch.set_facecolor('white')
fig.set_dpi(200)

plt.loglog(t_best_fit[time_mask], Teff_inf_best_fit[time_mask], label='Merritt et al Values')
plt.loglog(t_best_fit2[time_mask2], Teff_inf_best_fit2[time_mask2], label='MCMC Values')
plt.errorbar(t_obs, Teff_obs, yerr=Teff_obs_sigma, fmt='.', label='Observational Data')
plt.xlabel('t (days)')
plt.ylabel(r'T$_{eff}$ (MK)')
plt.title('KS 1731-260 Lightcurves')
plt.legend(loc='upper right', fontsize=6)
plt.show()
fig.savefig('lightcurves')
```

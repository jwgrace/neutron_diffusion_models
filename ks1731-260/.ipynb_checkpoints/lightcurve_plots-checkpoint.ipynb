```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
import os
import sys
from glob import glob
from ipywidgets import interactive

home_directory = os.getenv('HOME')
dStar_directory = home_directory + '/dStar/'
sys.path.append(dStar_directory + 'tools/')

import reader
```

```python
# Load parameter names and values for use in the widget
parameter_names = np.load('parameter_names.npy')
parameter_values = np.load('parameter_values.npy')
```

```python
# Load base model output
LOGS_directory_base = 'LOGS/'

f = reader.dStarReader('{}history.data'.format(LOGS_directory_base))

# time after outburts in days
t_base = f.data.time
# log(L) [ergs]
lg_L_base = f.data.lg_Lsurf
# log(T_eff) [K]
lg_Teff_base = f.data.lg_Teff
```

```python
# load models with neutron diffusion

# Load LOGS directories of neutron diffusion models
neutron_diffusion_LOGS_directories = sorted(glob('neutron_diffusion_LOGS/LOGS????/'))

num_models = len(neutron_diffusion_LOGS_directories)

# Array of arrays of luminosity and Teff values for each neutron diffusion model
lg_L_neutron_diffusion = np.zeros((num_models, len(lg_L_base)))
lg_Teff_neutron_diffusion = np.zeros((num_models, len(lg_Teff_base)))

# Iterate through each neutron diffusion model and load luminosity and temperature data.
# Interpolate the data to match the time outputs of the base lightcurve to compare.
for i, LOGS_directory in enumerate(neutron_diffusion_LOGS_directories):
    
    f = reader.dStarReader('{}history.data'.format(LOGS_directory))

    # Load arrays of time, log(L) and Teff
    t_model = f.data.time
    lg_L_model = f.data.lg_Lsurf
    lg_Teff_model = f.data.lg_Teff
    
    # Interpolate luminosity data
    interpolation = interpolate.interp1d(t_model, lg_L_model)
    lg_L_neutron_diffusion[i] = interpolation(t_base)
    # Interpolate Teff data
    interpolation = interpolate.interp1d(t_model, lg_Teff_model)
    lg_Teff_neutron_diffusion[i] = interpolation(t_base)
```

```python
# Load data from output file including observational data.
output_file = open('{}output'.format(LOGS_directory_base), 'r')

t_obs, Teff_model, Teff_min_obs, Teff_max_obs = np.loadtxt(output_file, skiprows=4, max_rows=8, unpack=True)

print(Teff_model)
```

```python
k_B = 8.617e-5
Teff_obs = np.array([103.2,88.9,75.5,73.3,71.0,66.0,70.3,63.1])/k_B*10**-6
Teff_obs_range_mean = (Teff_min_obs + Teff_max_obs)/2
Teff_obs_range_sigma = (Teff_max_obs - Teff_min_obs)/2

Teff_obs_sigma = np.array([1.7,1.3,2.2,2.3,1.8,4.5,2.1,2.1])/k_B*10**-6

chi2 = np.sum((Teff_model-Teff_obs)**2 / Teff_obs_sigma**2 )
print(chi2)

chi2 = np.sum((Teff_model-Teff_obs_range_mean)**2 / Teff_obs_range_sigma**2 )
print(chi2)
```

```python
c = 3e10
G = 6.67e-8
M_sun = 1.989e33
M_ns = 1.4*M_sun
R_ns = 1e6

redshift_factor = np.sqrt(1 - 2*G*M_ns/(R_ns*c**2))

Teff_base_MK = 10**lg_Teff_base*10**-6
Teff_base_MK_inf = Teff_base_MK*redshift_factor

print(redshift_factor)
```

```python
fig = plt.figure()
fig.patch.set_facecolor('white')
fig.set_dpi(200)

plt.plot(t_base, Teff_base_MK_inf)
plt.plot(t_obs, Teff_model, '.')
plt.show()
```

```python
interpolation = interpolate.interp1d(t_base, lg_Teff_base)
lg_Teff_base2 = interpolation(t_obs)
print(lg_Teff_base2/np.log10(Teff_model*10**6))
print(10**lg_Teff_base2/(Teff_model*10**6))
```

```python
%matplotlib inline

def plot_func(model_number):
    x = t_base
    y0 = lg_Teff_base
    y = lgL_neutron_diffusion[model_number]
    x_obs = t_obs
    
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
    
    """fig = plt.figure()
    fig.patch.set_facecolor('white')
    fig.set_dpi(200)"""

    ax1.plot(x, y0)
    ax1.plot(x, y, '--')
    ax1.set(ylim=(32.8, 34.2), xlabel='Time', ylabel='Log (L)')
    
    ax2.plot(x, np.zeros_like(x))
    ax2.plot(x, y - y0, '--')
    ax2.set(ylim=(-.05, .05), xlabel='Time', ylabel=r'Log (L) - Log (L$_0$)')
    
    fig.suptitle('{} = {}'.format(parameter_names[0], parameter_values[model_number]), fontsize=24)
    
    plt.show()

interactive_plot = interactive(plot_func, model_number=(0, len(parameter_values)-1))
output = interactive_plot.children[-1]
interactive_plot
```
